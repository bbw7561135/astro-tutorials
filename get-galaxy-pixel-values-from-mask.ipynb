## Use a SExtractor Segmentation Image to Get Galaxy Pixel Values

The goal is to find the pixel values associated with a galaxy.  We are going to use a sextractor segmentation image to identify the pixels.

```python
from astropy.io import fits
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

We will use two images.

* R-band image: pointing4-88353-R.fits
* SE segmentation: pointing4-88353-R-segmentation.fits

```python
image = fits.getdata('images/pointing4-88353-R.fits')
```

```python
mask = fits.getdata('images/pointing4-88353-R-segmentation.fits')
```

```python
plt.figure()
v1 = np.percentile(image,5)
v2 = np.percentile(image,99)
plt.imshow(image, origin='upper',vmin=v1, vmax=v2,cmap='gray')
plt.title('R-band Image')
plt.colorbar(fraction=0.08)

plt.figure()
v1 = np.percentile(mask,0)
v2 = np.percentile(mask,99)
plt.imshow(mask, origin='upper',vmin=v1, vmax=v2,cmap='gray')
plt.title('SE Segmentation Image')
plt.colorbar(fraction=0.08)
```

The galaxy is object number 1 in the segmentation image, so we want to keep the pixels that correspond to object 1 in the segmentation image.  

```python
galaxy_pixels = image[mask == 1]
```
